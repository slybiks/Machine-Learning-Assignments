```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os
path = os.getcwd() + '/auto-mpg.csv'
raw_data = pd.read_csv(path)
raw_data.head()
```

```python
data = ( raw_data - raw_data.mean() )/( raw_data.max()-raw_data.min())
data.head()
```

```python
data.shape
```

```python
cols = data.shape[1]
print(cols)
```

```python
data.insert(cols,'Ones',1)
data.head()
```

```python
cols = data.shape[1]
print(cols)
```

```python
x = data.iloc[:,1:cols]
y = data.iloc[:,:1]
```

```python
y.head()
```

```python
x = np.matrix(x)
y = np.matrix(y)
theta = np.matrix(np.array([0,0]))
x.shape, theta.shape, y.shape
```

```python
def computeError(x,y,theta):
    inner = np.power(((x * theta.T) - y), 2)
    return np.sum(inner)/(2 * len(x))
```

```python
computeError(x,y,theta)
```

```python
Learn_rate = 1
iters = 1000
def gradientDescent(x,y,theta,Learn_rate,iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error = (x * theta.T) - y
        
        for j in range(parameters):
            term = np.multiply(error,x[:,j])
            temp[0,j] = theta[0,j] - ((Learn_rate/len(x)) * np.sum(term))
            
            
        theta = temp
        cost[i] = computeError(x,y,theta)
        
    return theta,cost  

new_theta,cost = gradientDescent(x,y,theta,Learn_rate,iters)
print(new_theta,cost)
```

```python
Model_mpg = new_theta[0,0] + (new_theta[0,1]*x)
```

```python
import math
error = [np.power((b-a),2) for (a,b) in zip(Model_mpg,y)]
error0 = np.sum(error)
error1 = math.sqrt(error0)
error = (error0/len(y))*100
print("Error% = {}".format(error))
accuracy = 100 - error
print("Accuracy% = {}".format(accuracy))
```
